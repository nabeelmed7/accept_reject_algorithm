```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# defining f(x) (Target distribution)
def f(x): 
    t = (-x**2)/2
    f_x = np.exp(t)*((np.sin(6*x)**2)+(3*np.cos(x)**2)*(np.sin(4*x)**2)+1)
    return f_x
```

```python
# defining g(x) (Proposal distribution)
def g(x):
    t = (-x**2/2)
    g_x = np.exp(t)/np.sqrt(2*np.pi)
    return g_x
```

```python
# accept reject sampling
def accept_reject(n):
  X = np.random.randn(n)

  plt.rcParams["figure.figsize"] = [15.00, 3.50]
  plt.rcParams["figure.autolayout"] = True

  fig, axes = plt.subplots(1,3)
  print(sns.lineplot(x=X, y=f(X), ax=axes[0], label = 'f(x)').set(title='Distribution of f(x) and g(x) before scaling'))
  print(sns.lineplot(x=X, y=g(X),ax = axes[0],  label = 'g(x)'))

  M = np.max(f(X)/g(X)) # Using max to get an optimal value of M

  print(sns.lineplot(x=X, y=f(X), ax=axes[1], label = 'f(x)').set(title='Distribution of f(x) and M*g(x) after scaling'))
  print(sns.lineplot(x=X, y=M*g(X), ax = axes[1], label = 'M*g(x)'))

  u = np.random.uniform(0, M*g(X)) # picking a uniformly distributed random variable between 0 and M*g(x)
  accepted_samples = X[u <= f(X)] # accepting only those samples from X where u <= f(X)
  percent_of_samples_accpeted = len(accepted_samples)/n

  print(sns.lineplot(x=accepted_samples,y=f(accepted_samples),ax = axes[2]).set(title='Distribution of accepted samples'))
  print('---------------------------------------------------------------------------------------------------------------------------------------')
  
  return print(f'Total number of accepted samples = {len(accepted_samples)}\nPercent of accepted samples = {percent_of_samples_accpeted}')
```

```python
accept_reject(5000)
```
